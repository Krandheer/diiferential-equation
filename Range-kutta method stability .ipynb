```python
'''euler stability region contour/surface plot, we got stability region
abs(1+z+z^2/3 + z^3/6 + z^4/24)<1, on solving eqaution y'=ky, (where k belongs to complex number and y(0)=1)
'''

import matplotlib.pyplot as plt
import math
import numpy as np

xv=np.linspace(-5,2,1000)
yv=np.linspace(-5,1,1000)

# array which will store the value which lies in stability region
x=[]
y=[]

#search of stability region values
for i in xv:
    for j in yv:
        
        
        #converting abs(1+z+z^2/3 + z^3/6 + z^4/24)<1 into x and y and storing those point which satisfies the condition 
        if math.sqrt((i**4 + j**4 + 4*(i**3) - 6*i*i*j*j + 12*(i*i-j*j) - 12*i*j*j + 24*i + 24)**2
                    +(4*(i**3)*j*j + 12*i*i*j + 2*i*(j**3) - 2*i*j*j + 4*(j**3) + 24*i*j + 24*j)**2) <24:
            x.append(i)
            y.append(j)

#stability region plot
fig=plt.figure(figsize=(10,10))
plt.plot(x,y,'b')
plt.axis('equal')
plt.title('range kutta method of order 4 stability region plot', fontsize=14)
plt.xlabel('real lambda*h')
plt.ylabel('imaginary lambda*h')
plt.savefig('range_kutta4_fig.png')
```
