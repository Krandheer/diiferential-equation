```python
'''euler stability region contour plot, the stability region we got on solving eqaution y' = ky where
k belongs to complex number, and y(0) = 1, is abs(1+z) < 1 which is a cirle with center at (-1,0) and 
radius 1,so here are plotting for abs(1+z)=1 and we will consider the inner region, 
not boundary the stability region for euler method'''

import matplotlib.pyplot as plt
import math
import numpy as np

xv=np.linspace(-3,1,2000)
yv=np.linspace(-2,2,2000)

#array which will store the values which lies in the stability region
x=[]
y=[]

#search for stability region values
for i in xv:
    for j in yv:
        
        # converting the fucntion abs(1+z)<=1 in x and y coordinate and then iterating for x and y
        if math.sqrt((i+1)**2 + j**2) <1: 
            x.append(i)
            y.append(j)

# plot of stability region 
plt.plot(x,y,'b')
plt.axis('equal')
plt.xlabel('x-axis, real lambda*h')
plt.ylabel('y-axis, imaginary lambda*h')
plt.title('euler method stability region plot', fontsize =14)
```
