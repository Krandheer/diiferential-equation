```python
'''euler stability region contour plot, we got stability region on solving
abs(1+z+z^2)<1 which is the region outside of the cirle with center at (1,0) and radius 1,
'''

import matplotlib.pyplot as plt
import math
import numpy as np

xv=np.linspace(0,2,500)
yv=np.linspace(-1,1,500)

# array which will store the value which lies in stability region
x=[]
y=[]

#search of stability region values
for i in xv:
    for j in yv:
         
        #converting abs(1+z+z^2)<1 into x and y and storing those point which satisfies the condition
        if math.sqrt((1+i*i + i-j*j)**2 + (2*i*j+j)**2) <1:
            x.append(i)
            y.append(j)

#stability region plot
fig=plt.figure(figsize=(10,10))
plt.plot(x,y,'b')
plt.axis('equal')
plt.title('backward euler method stability region plot', fontsize=14)
plt.xlabel('real lambda*h')
plt.ylabel('imaginary lambda*h')
plt.savefig('fig.png')

```
